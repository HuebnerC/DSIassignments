```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1000, size=50)
y = np.random.randint(1000, size=50)
```

```python
(x, y)
```

```python
fig, ax = plt.subplots()
color = np.where((x + y) % 2 == 0, 'r', 'b')
ax.scatter(x, y, c=color, label=color)
ax.set_xlabel('x values')
ax.set_ylabel('y values')
ax.legend()
```

## Two functions

```python

x = np.linspace(0,5,100)

ax[0, 1].plot(x,3 * x + 0.5, color='green', ls="--", label="$P_2(x)$")
ax[0, 1].plot(x,5 * np.sqrt(x), color='blue', ls=":", label="$P_3(x)$")

ax[0,1].legend();
```

### Graph with Log Scale

```python
x = np.linspace(0,5,100)

ax = fig.add_subplot(1,1, 1)
ax.set_xscale('log')
#Plot something
ax.plot(x,x, color='red', ls="-", label="$P_1(x)$")
ax.plot(x,3 * x + 0.5, color='green', ls="--", label="$P_2(x)$")
ax.plot(x,5 * np.sqrt(x), color='blue', ls=":", label="$P_3(x)$")

ax.legend();

```

## Importing data with numpy

```python
cols = [2, 5, 8, 11, 14, 17]
filepath = 'data/bay_area_bikeshare/201402_weather_data_v2.csv'
weather = np.loadtxt(filepath, delimiter=',', skiprows=1, usecols=cols)
weather
```

```python
fig, ax = plt.subplots()

ax.scatter(weather[:, 0], weather[:, 1])
ax.set_xlabel('mean_temperature')
ax.set_ylabel('mean_dew_point_f')
```

```python
# Checking to see if there is a correlation between windspeed and pressure - i.e. a cold front blowing in
# used .set_xlim and ylim, decided it made the graph too clustered if (0,0) origin
fig, ax = plt.subplots()
ax.scatter(weather[:, 5], weather[:, 3])
ax.set_xlabel('mean_wind_speed_m_p_h')
ax.set_ylabel('mean_sea_level_pressure_in')
```
